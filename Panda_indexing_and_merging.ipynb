# Merging dataframes

```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame(dict(id=range(4), age=np.random.randint(18, 31, size=4)))
df2 = pd.DataFrame(dict(id=list(range(5))+list(range(5)), score=np.random.random(size=10)))

print(df1)
print(df2)
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```

```python
pd.merge(df1, df2, how = 'outer')
```
